```python
import pandas as pd
import matplotlib.pyplot as plt


plt.rc('font', family='Malgun Gothic') 

file_path = '출동현황.csv'
Cause = pd.read_csv(file_path, encoding='utf-8')

value_counts = Cause['시설장소'].value_counts()
print(value_counts)

plt.figure(figsize=(10, 6))
plt.title('화재장소', fontsize=20, weight='bold')
plt.xlabel('빈도수', fontsize=15)
plt.ylabel('')

value_counts.plot.barh(color='red', edgecolor='black')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
# plt.tight_layout()

for index, value in enumerate(value_counts):
    plt.text(value + 0.5, index, str(value), fontsize=12, va='center')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = '출동현황.csv'
Cause = pd.read_csv(file_path, encoding='utf-8')
#print(Cause.columns)
value_counts = Cause['시설장소'].value_counts()
print(value_counts)

plt.figure(figsize=(7, 7))  
plt.title('화재장소', fontsize=20) 

value_counts.plot.pie(
    autopct='%.2f%%', colors=plt.cm.Paired.colors,
    textprops={'fontsize': 10, 'weight': 'bold'})

plt.ylabel('')
plt.show()
```
